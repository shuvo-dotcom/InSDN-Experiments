```python
from datasets import load_dataset, DatasetDict
from sklearn.model_selection import train_test_split
import pandas as pd

# Load the Hugging Face dataset
dataset = load_dataset("Suvo696/attack", split="train")

# Convert to pandas DataFrame for processing
df = pd.DataFrame(dataset)
```

```python
df
```

```python
# Clean duplicate class names if any (e.g., repeated DDoS)
df["Label"] = df["Label"].str.strip().str.upper()
df = df.drop_duplicates()

# Split into train and test sets with stratification
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["Label"],
    random_state=42
)
```

```python
import os

# Assuming train_df and test_df are already in memory from previous steps
output_dir = "prepared_insdn_data"
os.makedirs(output_dir, exist_ok=True)

train_df.to_csv(os.path.join(output_dir, "train.csv"), index=False)
test_df.to_csv(os.path.join(output_dir, "test.csv"), index=False)

print("✅ Train and test datasets saved successfully!")
```
